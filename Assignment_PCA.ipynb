```python
from sklearn.decomposition import PCA
pca = PCA()
```

```python
pc = pca.fit_transform(wine)
pc
```

```python
pca.components_
```

```python
# The amount of variance that each PCA has
var = pca.explained_variance_ratio_
var
```

```python
# Cummulative variance of each PCA
Var = np.cumsum(np.round(var,decimals= 4)*100)
Var
```

```python
plt.plot(Var,color="blue");
```

```python
df = pd.concat([df['Type'],pd.DataFrame(pc[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
df
```

# Heirarchial 

```python
X = df.values
X
```

```python
# Create Dendrograms
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(10, 7))  

dendograms=sch.dendrogram(sch.linkage(X,method = 'ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
Y = cluster.fit_predict(X)
```

```python
Y_new = pd.DataFrame(Y)
```

```python
Y_new[0].value_counts()
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(X[:,1], X[:,2], c=Y_new, cmap='rainbow') 
```

# k mean clustering

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range (1,11):
    km = KMeans(n_clusters = i)
    km.fit_predict(X)
    wcss.append(km.inertia_)
print(wcss)
```

```python
import matplotlib.pyplot as plt
plt.scatter(range(1,11),wcss)
plt.plot(range(1,11),wcss)
```

```python
km = KMeans(n_clusters = 3)
Y = km.fit_predict(X)
Y
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(X[:,1], X[:,2], c=Y, cmap='rainbow')  
```
